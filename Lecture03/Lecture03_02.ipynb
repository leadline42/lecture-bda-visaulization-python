```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
file_path = './Data02/'
file_list = os.listdir(file_path)

df_happy_health = pd.read_excel(file_path+file_list[0])
df_happy_econo = pd.read_excel(file_path+file_list[1])
df_happy_relation = pd.read_excel(file_path+file_list[2])
df_happy_edu = pd.read_excel(file_path+file_list[3])
df_happy_life = pd.read_excel(file_path+file_list[4])
df_happy_safe = pd.read_excel(file_path+file_list[5])
df_happy_leisure = pd.read_excel(file_path+file_list[6])
df_happy_environ = pd.read_excel(file_path+file_list[7])
```

```python
city = df_happy_safe['시도'].unique()
df_happy = pd.DataFrame({'시도':city})
df_happy_sort = df_happy.sort_values(by='시도').reset_index(drop=True)
df_happy_sort
```

```python
df_life_mean = pd.DataFrame(df_happy_life['삶의 만족도'].groupby(by=df_happy_life['시도']).mean()).reset_index()
df_life_mean
```

```python
pd.concat([df_happy_sort, df_life_mean.iloc[:,1]], axis=1)
```

```python
df_happy_merge = df_happy.merge(df_life_mean, how='left', on='시도')
df_happy_merge
```

```python
df_health_mean = pd.DataFrame(df_happy_health['평균'].groupby(by=df_happy_health['시도']).mean()).reset_index()
df_health_mean.columns = ['시도', '건강']
df_health_mean
```

```python
df_happy_merge = df_happy_merge.merge(df_health_mean, how='left', on='시도')
df_happy_merge
```

```python
df_safe_mean = pd.DataFrame(df_happy_safe['평균'].groupby(by=df_happy_safe['시도']).mean()).reset_index()
df_safe_mean.columns = ['시도', '안전']
df_safe_mean
```

```python
df_happy_merge = df_happy_merge.merge(df_safe_mean, how='left', on='시도')
df_happy_merge
```

```python
df_environ_mean = pd.DataFrame(df_happy_environ['평균'].groupby(by=df_happy_environ['시도']).mean()).reset_index()
df_environ_mean.columns = ['시도', '환경']
df_environ_mean
```

```python
df_happy_merge = df_happy_merge.merge(df_environ_mean, how='left', on='시도')
df_happy_merge
```

```python
df_econo_mean = pd.DataFrame(df_happy_econo['평균'].groupby(by=df_happy_econo['시도']).mean()).reset_index()
df_econo_mean.columns = ['시도', '경제']
df_econo_mean
```

```python
df_happy_merge = df_happy_merge.merge(df_econo_mean, how='left', on='시도')
df_happy_merge
```

```python
df_edu_mean = pd.DataFrame(df_happy_edu['평균'].groupby(by=df_happy_edu['시도']).mean()).reset_index()
df_edu_mean.columns = ['시도', '교육']
df_edu_mean
```

```python
df_happy_merge = df_happy_merge.merge(df_edu_mean, how='left', on='시도')
df_happy_merge
```

```python
df_relation_mean = pd.DataFrame(df_happy_relation['평균'].groupby(by=df_happy_relation['시도']).mean()).reset_index()
df_relation_mean.columns = ['시도', '관계']
df_relation_mean
```

```python
df_happy_merge = df_happy_merge.merge(df_relation_mean, how='left', on='시도')
df_happy_merge
```

```python
df_leisure_mean = pd.DataFrame(df_happy_leisure['평균'].groupby(by=df_happy_leisure['시도']).mean()).reset_index()
df_leisure_mean.columns = ['시도', '여가']
df_leisure_mean
```

```python
df_happy_merge = df_happy_merge.merge(df_leisure_mean, how='left', on='시도')
df_happy_merge
```
