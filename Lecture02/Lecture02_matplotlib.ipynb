## 2. 맷플롯립(Matplotlib) 라이브러리

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Coding Style
- https://matplotlib.org/stable/tutorials/introductory/quick_start.html#coding-styles

Matplotlib's documentation and examples use both the OO and the pyplot styles. <b>In general, we suggest using the OO style</b>, particularly for complicated plots, and functions and scripts that are intended to be reused as part of a larger project. <b>However, the pyplot style can be very convenient for quick interactive work.</b>

```python
# OO(Object-Oriented)-Style
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
# pyplot-Style
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
# DataFrame을 활용한 시각화
my_score = [[50, 90, 95], [60, 75, 100], [75, 85, 90], [85, 75, 90], [95, 80, 85], [90, 85, 80], [95, 80, 75]]
subject = ['kor', 'math', 'eng']
df = pd.DataFrame(my_score, columns=subject)
```

```python
# Pandas 자체 Plot 기능: 빠르게 시각화가 가능
df.plot.line(), df.plot.area(), df.plot.box()
plt.show()
```

```python
# matplotlib 활용
fig, axs = plt.subplots(3,1)
axs[0].plot(df, label=subject)
axs[1].stackplot(df.index, df.values.T, labels=subject)
axs[2].boxplot(df)
ax.legend()

plt.show()
```

## 선 그래프 그리기

```python
# 미세먼지 데이터 불러오기
df_rawdata_fine_dust = pd.read_excel('./Data/fine_dust.xlsx', index_col='area')
df_rawdata_fine_dust.info()
df_rawdata_fine_dust.head()
```

```python
# 미세먼지 선 그래프 그리기
fig, ax = plt.subplots()

fig.set_size_inches(15,4)
ax.plot(df_rawdata_fine_dust[2018], color='b', marker='o')
ax.set_xlabel('area')
ax.set_ylabel('micrometer')
ax.set_title('2018 Fine Dust Line Graph')
ax.grid()

plt.show()
```

```python
# 2016~2019년 지역별 미세먼지 선 그래프
fig, ax = plt.subplots()

fig.set_size_inches(15,4)

for year in range(2016, 2020):
  ax.plot(df_rawdata_fine_dust[year], marker='s', label=year)
ax.set_xlabel('area')
ax.set_ylabel('micrometer')
ax.set_title('2016~2019 Fine Dust Line Graph')
ax.grid()

plt.show()
```

## 바 그래프 그리기

```python
# 2017년 미세먼지 세로 막대 그래프 그리기
fig, ax = plt.subplots()

fig.set_size_inches(15,4)
ax.bar(df_rawdata_fine_dust.index, df_rawdata_fine_dust[2017], color='g')
ax.set_ylim(35,55)
ax.set_xlabel('area')
ax.set_ylabel('micrometer')
ax.set_title('2017 Fine Dust Bar Graph')
ax.grid()

plt.show()
```

```python
# 2016~2019년 경기, 인천, 부산, 대구 지역별 미세먼지 세로 막대 그래프 그리기
index = np.arange(4)

fig, ax = plt.subplots()

fig.set_size_inches(15,4)

df_mod00_fine_dust = df_rawdata_fine_dust.loc['Gyeonggi':'Daegu', 2016:2019]
for year in range(2016, 2020):
  ax.bar(index, df_mod00_fine_dust[year], width=0.2, label=year)
  index = index + 0.2

ax.set_ylim(35,55)
ax.set_xlabel('area')
ax.set_ylabel('micrometer')
ax.set_xticks(index-0.5, ['Gyeonggi', 'Incheon', 'Busan', 'Daegu'])
ax.set_title('2016~2019 Fine Dust Group Bar Graph')
ax.legend()

plt.show()
```

```python
# 2016~2019년 지역별 미세먼지 누적 가로 막대 그래프 그리기
index = np.arange(4)

fig, ax = plt.subplots()

fig.set_size_inches(10,4)

df_mod00_fine_dust = df_rawdata_fine_dust.loc['Gyeonggi':'Daegu', 2016:2019]
for year in range(2016, 2020):
  ax.barh(index, df_mod00_fine_dust[year], label=year)

ax.set_xlim(30,55)
ax.set_ylabel('area')
ax.set_xlabel('micrometer')
ax.set_yticks(index, ['Gyeonggi', 'Incheon', 'Busan', 'Daegu'])
ax.set_title('2016~2019 Fine Dust Group Barh Graph')
ax.legend()

plt.show()
```

## 산점도 그래프 그리기

```python
df_rawdata_health = pd.read_excel('./Data/health_screenings_2020_1000ea.xlsx')
df_rawdata_health.info()
df_rawdata_health.head()
```

```python
# 키와 몸무게 산점도 그래프
heightdata = df_rawdata_health['height']
weightdata = df_rawdata_health['weight']

fig, ax = plt.subplots()

fig.set_size_inches(10,4)

ax.scatter(heightdata, weightdata)
ax.set_xlabel('height')
ax.set_ylabel('weight')
ax.set_title('2020 Health Screenings Scatter Graph')
ax.grid()

plt.show()
```

```python
# 수축기 혈압, 이완기 혈압, 공복 혈당 그룹 산점도 그래프
systolic_data = df_rawdata_health['systolic']
diastolic_data = df_rawdata_health['diastolic']
blood_sugar_data = df_rawdata_health['blood_sugar']

fig, ax = plt.subplots()

fig.set_size_inches(10,6)

ax.scatter(systolic_data, diastolic_data, color='r', edgecolor='w', label='systolic*diastolic')
ax.scatter(systolic_data, blood_sugar_data, color='g', edgecolor='w', label='systolic*blood_sugar')
ax.scatter(diastolic_data, blood_sugar_data, color='b', edgecolor='w', label='diastolic*blood_sugar')

ax.set_xlim(40,180)
ax.set_ylim(30,300)
ax.set_title('2020 Health Screenings Group Scatter Graph')
ax.legend()
ax.grid()

plt.show()
```

## 히스토그램 그래프 그리기

```python
# 남성 키 히스토그램
female_data = df_rawdata_health.loc[df_rawdata_health.gender==2,['gender','height']]

fig, ax = plt.subplots()

fig.set_size_inches(10,6)

ax.hist(female_data['height'], bins=7, label='Female')
ax.set_xlim(130,180)
ax.set_xlabel('height')
ax.set_ylabel('frequency')
ax.set_title('2020 Health Screenings Female Height Histogram')
ax.legend()
ax.grid()

plt.show()
```

```python
# 남성과 여성 키 집단 히스토그램
male_data = df_rawdata_health.loc[df_rawdata_health.gender==1,['gender','height']]
female_data = df_rawdata_health.loc[df_rawdata_health.gender==2,['gender','height']]

fig, ax = plt.subplots()

fig.set_size_inches(10,6)

ax.hist(male_data['height'], bins=7, alpha=0.5, label='Male')
ax.hist(female_data['height'], bins=7, alpha=0.5, label='Female')
ax.set_xlim(130,195)
ax.set_xlabel('height')
ax.set_ylabel('frequency')
ax.set_title('2020 Health Screenings Male & Female Height Group Histogram')
ax.legend()
ax.grid()

plt.show()
```

```python
# 남성과 여성 키 집단 누적 히스토그램
height = np.array([male_data['height'], female_data['height']], dtype=object)

fig, ax = plt.subplots()

fig.set_size_inches(10,6)

ax.hist(height, bins=6, label=['Male','Female'], stacked=True)
ax.set_xlim(130,195)
ax.set_xlabel('height')
ax.set_ylabel('frequency')
ax.set_title('2020 Health Screenings Male & Female Height Stacked Histogram')
ax.legend()

plt.show()
```

## 상자수염 그래프 그리기

```python
# 남성과 여성 키 상자수염 그래프
height = np.array([male_data['height'], female_data['height']], dtype=object)

fig, ax = plt.subplots()

fig.set_size_inches(8,5)

ax.boxplot(height, labels=['Male','Female'])
ax.set_ylim(130,200)
ax.set_xlabel('gender')
ax.set_ylabel('height')
ax.set_title('2020 Health Screenings Male & Female Height Box Plot')

plt.show()
```

```python
# 남성과 여성 키 가로 상자수염 그래프
fig, ax = plt.subplots()

fig.set_size_inches(8,5)

ax.boxplot(height, labels=['Male','Female'], vert=False)

ax.set_xlim(130,200)
ax.set_ylabel('gender')
ax.set_xlabel('height')
ax.set_title('2020 Health Screenings Male & Female Height Box Plot')

plt.show()
```
