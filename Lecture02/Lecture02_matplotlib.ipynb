### Coding Style
- https://matplotlib.org/stable/tutorials/introductory/quick_start.html#coding-styles

Matplotlib's documentation and examples use both the OO and the pyplot styles. <b>In general, we suggest using the OO style</b>, particularly for complicated plots, and functions and scripts that are intended to be reused as part of a larger project. <b>However, the pyplot style can be very convenient for quick interactive work.</b>

```python
# OO(Object-Oriented)-Style
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
# pyplot-Style
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
np.random.seed(42)

dt = 0.01
t = np.arange(0, 30, dt)

nse1 = np.random.randn(len(t))
nse2 = np.random.randn(len(t))

s1 = np.sin(2*np.pi*10*t) + nse1
s2 = np.sin(2*np.pi*10*t) + nse2

fig, axs = plt.subplots(2, 1)
fig.set_size_inches(10,8)

axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel("Time")
axs[0].set_ylabel("s1 and s2")
axs[0].grid()

cxy, f = axs[1].cohere(s1, s2, 256, 1. / dt)
axs[1].set_ylabel("Coherence")

fig.tight_layout()
plt.show()
```

## 산점도 그래프 그리기

```python
df_rawdata_health = pd.read_excel('./Data/health_screenings_2020_1000ea.xlsx')
df_rawdata_health.info()
df_rawdata_health.head()
```

```python
# 키와 몸무게 산점도 그래프
font = {'family': 'serif',
        'color':  'darkblue',
        'weight': 'normal',
        'size': 18,
        }

heightdata = df_rawdata_health['height']
weightdata = df_rawdata_health['weight']

fig, ax = plt.subplots()
fig.set_size_inches(10,4)

ax.scatter(heightdata, weightdata, color='r', marker='o')
ax.set_xlim(130, 195)
ax.set_xlabel('height', fontdict=font)
ax.set_ylabel('weight', fontdict=font)
ax.set_title('2020 Health Screenings Scatter Graph', fontdict=font)
ax.grid()

plt.show()
```

```python
# 수축기 혈압, 이완기 혈압, 공복 혈당 그룹 산점도 그래프
font = {'family': 'serif',
        'color':  'darkblue',
        'weight': 'normal',
        'size': 18,
        }

systolic_data = df_rawdata_health['systolic']
diastolic_data = df_rawdata_health['diastolic']
blood_sugar_data = df_rawdata_health['blood_sugar']

fig, ax = plt.subplots()
fig.set_size_inches(10,6)

ax.scatter(systolic_data, diastolic_data, color='r', edgecolor='w', label='systolic*diastolic')
ax.scatter(systolic_data, blood_sugar_data, color='g', edgecolor='w', label='systolic*blood_sugar')
ax.scatter(diastolic_data, blood_sugar_data, color='b', edgecolor='w', label='diastolic*blood_sugar')

ax.set_xlim(40,180)
ax.set_ylim(30,300)
ax.set_title('2020 Health Screenings Group Scatter Graph', fontdict=font)
ax.legend()
ax.grid()

plt.show()
```

```python
np.random.seed(24)

x = np.random.randn(1000)
y = np.random.randn(1000)
```

```python
def scatter_hist(x, y, ax, ax_histx, ax_histy):
    ax_histx.tick_params(axis='x', labelbottom=False)
    ax_histy.tick_params(axis='y', labelleft=False)

    ax.scatter(x, y)

    binwidth = 0.25
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth)+1) * binwidth

    bins = np.arange(-lim, lim + binwidth, binwidth)
    ax_histx.hist(x, bins=bins)
    ax_histy.hist(y, bins=bins, orientation='horizontal')
```

```python
fig = plt.figure(figsize=(6, 6))

gs = fig.add_gridspec(2, 2,  width_ratios=(4, 1), height_ratios=(1, 4),
                      left=0.1, right=0.9, bottom=0.1, top=0.9,
                      wspace=0.05, hspace=0.05)

ax = fig.add_subplot(gs[1, 0])
ax_histx = fig.add_subplot(gs[0, 0], sharex=ax)
ax_histy = fig.add_subplot(gs[1, 1], sharey=ax)

scatter_hist(x, y, ax, ax_histx, ax_histy)
```
